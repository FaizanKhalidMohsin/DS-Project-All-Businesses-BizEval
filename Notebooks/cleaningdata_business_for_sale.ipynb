```python
#This code is for https://canada.businessesforsale.com/
import pandas as pd
import numpy as np
import re
```

```python
#This section loads the raw data and the clean data file paths. The raw data is read into a DataFrame for processing.
rawData = pd.read_csv("/content/Browse AI - 250 Rows - Dec 18 2024.csv")
raw_data_path = '/content/Browse AI - 250 Rows - Dec 18 2024.csv'
clean_data_path = '/content/cleanData.csv'
raw_data = pd.read_csv(raw_data_path)
```

```python
# Column Mapping
# Defines how columns in the raw data should be renamed to match the format of the cleaned data.
column_mapping = {
    "Position": "Position",  # Keeps the Position column as is.
    "Business Name": "Business Name",  # Renames Business Title to Business Title.
    "Location": "Location",  # Keeps the Location column as is.
    "Description": "Description",  # Keeps the Description column as is.
    "Asking Price": "Asking Price",  # Keeps the Asking Price column as is.
    "Cash Flow": "Cash Flow",  # Keeps the Cash Flow column as is.
    "Business URL": "Listing URL",  # Renames Business URL to Listing URL.
    "Image URL": "Image",  # Renames Image URL to Image.
}
# Rename columns
# Applies the column mapping to rename the raw data columns accordingly.
cleaned_data = raw_data.rename(columns=column_mapping)
```

```python
#General Data Cleaning
# Drop rows that are empty in the Business Title field. Check how many rows remain.
rawData.dropna(subset=["Business Name"]).shape
print('Rows removed', rawData.shape[0] - rawData.dropna(subset=["Business Name"]).shape[0])
# Check if all the rows that were retained were correct.
rawData.dropna(subset=["Business Name"])["Business Name"].values
```

```python
# Drop unnecessary columns
# Removes columns that are not required in the cleaned data.
columns_to_drop = [ "Image Alt Text", "Status"]
cleaned_data = cleaned_data.drop(columns=columns_to_drop)
```

```python
# Convert monetary values to numeric or retain ranges
# Processes monetary columns (e.g., Asking Price and Cash Flow) to clean up values and handle ranges properly.
def process_monetary_value(value):
    value = str(value).strip().upper()
    if not value or value == "":  # Handle empty strings or None
        return ""
    if "-" in value:  # Handle ranges
        range_parts = value.split("-")
        if len(range_parts) == 2:
            start = re.sub(r"[^0-9.]", "", range_parts[0])
            end = re.sub(r"[^0-9.]", "", range_parts[1])
            if "K" in value:
                return f"{int(float(start) * 1_000)}-{int(float(end) * 1_000)}"
            elif "M" in value:
                return f"{int(float(start) * 1_000_000)}-{int(float(end) * 1_000_000)}"
            return f"{int(float(start))}-{int(float(end))}"
    if "K" in value:  # Handle values with 'K'
        return int(float(re.sub(r"[^0-9.]", "", value)) * 1_000)
    elif "M" in value:  # Handle values with 'M'
        return int(float(re.sub(r"[^0-9.]", "", value)) * 1_000_000)
    else:
        # Use re.sub to keep only numeric values
        cleaned_value = re.sub(r"[^0-9.]", "", value)
        return int(float(cleaned_value)) if cleaned_value else ""

# Process Cash Flow column to blank out non-numerical values
def process_cashflow(value):
    try:
        # Attempt to process as a monetary value
        return process_monetary_value(value)
    except ValueError:
        # If not numerical, return empty string
        return ""


for column in ["Asking Price"]:
    cleaned_data[column] = cleaned_data[column].apply(process_monetary_value)

cleaned_data["Cash Flow"] = cleaned_data["Cash Flow"].apply(process_cashflow)

for column in ["Revenue"]:
    cleaned_data[column] = cleaned_data[column].apply(process_monetary_value)
```

```python
# Process Location column to extract city and province
# Keeps only the city and province from the Location column.
def process_location(value):
    match = re.search(r"([A-Za-z\s]+),\s*([A-Za-z\s]+)", value)
    if match:
        city = match.group(1).strip()
        province = match.group(2).strip()
        if province.lower() == "ontario":
            province = "ON"
        return f"{city}, {province}"
    return ""

cleaned_data["Location"] = cleaned_data["Location"].apply(process_location)
```

```python
# Process Cash Flow column to blank out non-numerical values
def process_cashflow(value):
    try:
        # Attempt to process as a monetary value
        return process_monetary_value(value)
    except ValueError:
        # If not numerical, return NaN
        return np.nan

for column in ["Asking Price"]:
    cleaned_data[column] = cleaned_data[column].apply(process_monetary_value)

cleaned_data["Cash Flow"] = cleaned_data["Cash Flow"].apply(process_cashflow)
```

```python
# Add derived fields if necessary
# Calculate the P/E Ratio by dividing Asking Price by Cash Flow, where applicable.
def calculate_pe_ratio(row):
    try:
        if isinstance(row["Cash Flow"], (int, float)) and row["Cash Flow"] > 0:
            return round(row["Asking Price"] / row["Cash Flow"], 1)
        return ""
    except:
        return ""

cleaned_data["P/E Ratio"] = cleaned_data.apply(calculate_pe_ratio, axis=1)
```

```python
# Remove rows with missing essential fields
# Drops rows where either Business Title or Asking Price is missing, as these are critical fields.
cleaned_data = cleaned_data.dropna(subset=["Business Name", "Asking Price"])
```

```python
# Save cleaned data
# Exports the cleaned data to a CSV file for further use.
output_path = '/content/cleaned_data_output.csv'
cleaned_data.to_csv(output_path, index=False)

# Confirmation message
print(f"Cleaned data saved to {output_path}")
```
